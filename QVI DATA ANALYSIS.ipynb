```python
QVI['DATE'] = pd.to_datetime(QVI['DATE'])
```

```python
QVI['DAY'] = QVI['DATE'].dt.day
QVI['MONTH'] = QVI['DATE'].dt.month
QVI['YEAR'] = QVI['DATE'].dt.year
QVI
```

```python
pd.options.display.max_rows = 10000
```

```python
QVI.isna().sum()
```

```python
QVI.head(100).sort_values(by = 'DATE')
```

```python
MONTHLY_SALES = QVI.groupby(['YEAR', 'MONTH'])['TOT_SALES'].sum().reset_index()
MONTHLY_SALES
```

```python
QVI.STORE_NBR.nunique()
```

# DECEMBER, 2018 HAD THE MOST SALES

```python
MAX_SALES_MONTH = MONTHLY_SALES.loc[MONTHLY_SALES['TOT_SALES'].idxmax()]
MAX_SALES_MONTH
```

```python
MON_SALES_18 = MONTHLY_SALES.loc[MONTHLY_SALES['YEAR'] == 2018]
MON_SALES_18.head()
```

```python
MON_SALES_19 = MONTHLY_SALES.loc[MONTHLY_SALES['YEAR'] == 2019]
MON_SALES_19.head()
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 7))
ax[0].bar(MON_SALES_18['MONTH'], MON_SALES_18['TOT_SALES'], color = 'black')
ax[1].bar(MON_SALES_19['MONTH'], MON_SALES_19['TOT_SALES'], color = 'black')

ax[0].set(title = '2018 MONTHLY SALES')
ax[1].set(title = '2019 MONTHLY SALES')

ax[0].set(xlabel = 'MONTH')
ax[0].set(ylabel = 'PRICE')

ax[1].set(xlabel = 'MONTH')
ax[1].set(ylabel = 'PRICE')
```
